```python
# ===================================================================
#                       1) LOAD

from ourpythonlibwithclasses.loading import loading_functions

path = "/content/drive/My Drive/google_colab/sample_diabetes_mellitus_data.csv"
test = loading_functions.Load_Split(path)

# Inspect
test.df
test.train
test.test
```

```python
# ===================================================================
#                       2) PREPROCESS

from ourpythonlibwithclasses.preprocessing import preprocessing_functions

#Define rows and cols to be dropped NA and filled NA, respectively.        
rows = ['age', 'gender', 'ethnicity']
cols  = ['height', 'weight']

# Drop rows w. NA on training data
testing = preprocessing_functions.PreProcessor(test.train)
testing.clean_rows(rows)
test_clean = testing.cleaned
print(test_clean)

 
# Fill cols w. NA on training data (w. dropped rows w. NA)
test_clean = preprocessing_functions.PreProcessor(test_clean)   
test_clean.fill_na_mean(cols)
test_fill = test_clean.filled
print(test_fill)


# Drop columns that are not int or float (to ease standardizing and polynomial transforming)
test_fill.dtypes
test_numeric = test_fill.select_dtypes(include = ['float', 'int'])
test_numeric.dtypes

# Standardizing
test_clean = preprocessing_functions.Standardize(test_numeric)
test_clean.feat_standardize()
test_scaled = test_clean.scaled
print(test_scaled)

# Polynomial transforming
#Drop cols with na
test_scaled = test_scaled.dropna()
test_scaled = preprocessing_functions.Standardize(test_scaled)
test_scaled.feat_poly()
print(test_scaled.polynized)
```

```python
# ===================================================================
#                       3) MODELLING
from ourpythonlibwithclasses.modelling import modelling_functions

model = modelling_functions.Model(test_scaled.polynized.drop('weight', axis = 1), test_scaled.polynized['weight'])
model.train()
model.predict()
```
